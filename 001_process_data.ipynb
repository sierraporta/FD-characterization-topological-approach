```python
import numpy as np
import pandas as pd
import scienceplots  # noqa: F401
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.style.use(["science", "nature"])
plt.rcParams.update({"font.size": 12, "xtick.labelsize": 12, "ytick.labelsize": 12, 
                     "axes.labelsize": 12,"legend.fontsize": 12})
#date = "2024-05-10"
#fixed_station = "calg"
```

### Loading

```python
def load_data(date):
    df=pd.read_csv("Datas/all{}.txt".format(date),delimiter=";")
    df=df.replace("   null",np.nan)
    df["DATETIME"]=pd.to_datetime(df["DATETIME"])
    df.set_index("DATETIME",inplace=True)
    df.index.name=None
    return df
```

```python
df20240510 = load_data(20240510)
df20240510.head(2)
```

### Plotting

```python
def plot_series(date,station,save=True):
    df=load_data(date)
    plt.figure(figsize=(14,3))
    ax=df[station].plot()
    plt.title("Event {} - Station {}".format(date,station))
    plt.xticks(rotation=0, ha='right'); plt.legend()
    plt.grid(alpha=0.3); plt.ylabel(None); plt.yticks(minor=True)
    if save:
        plt.savefig("Figures/CRcounts_{}_{}.png".format(station,date),dpi=300)
    return plt.show()    
```

```python
plot_series(20240510,"CALG")
```
